```python
# number of words/elements in post
def word_elements(post):
    return len(post.split(" "))

# word density
def word_den(post):
    den = len(post.split(" ")) / len(post)
    return den
```

```python
train['char_count'] = train['text'].apply(len)
test['char_count'] = test['text'].apply(len)
# number of words/elements in a post
train['word_count'] = train['text'].apply(word_elements)
test['word_count'] = test['text'].apply(word_elements)
# word density
train['word_den'] = train['text'].apply(word_den)
test['word_den'] = test['text'].apply(word_den)
```

```python
# Creating a single string contain tweets for the wordcloud
def corpus(df,column):
    return_string = ''
    
    for i in range(len(df[column])):
        return_string = return_string + ' ' + df[column].iloc[i]
    
    return return_string
```

```python
%%time
# split into train and test 
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.20, 
                                                    random_state=12)
```

```python
count = CountVectorizer(max_features=None,ngram_range=(1,1),min_df=1,
                       lowercase=True,
                       
                       ) # (max_features=None,ngram_range=(1,1),min_df=1)
```

```python
X_train = count.fit_transform(X_train.text)
X_test = count.transform(X_test.text)
```

```python
nb = MultinomialNB(alpha=0.2) # 0.2
```

```python
%%time
nb.fit(X_train, y_train)
```

```python
# train f1 score
pred = nb.predict(X_train)
print('train - ', round(f1_score(y_train, pred,average='micro'),4))

# test
pred = nb.predict(X_test)
print('test - ', round(f1_score(y_test, pred,average='micro'),7))
```

```python
%%time
# Classification report test set
print(classification_report(y_test, pred))
```

```python
# Get predictions on the test set 
y_hat = nb.predict_proba(X_test)
# Calculate the log loss (this was imported from sklearn above somewhere) 
print("The log loss error for our model is: ", log_loss(y_test, y_hat))
```
