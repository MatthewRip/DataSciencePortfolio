## Netflix Time Series Visualization

Dataset - https://www.kaggle.com/akpmpr/updated-netflix-stock-price-all-time

```python
# dependencies
import pandas as pd
import numpy as np

# Visualisations
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load Data
df = pd.read_csv('netflix.csv')
```

## Data Processing

```python
# convert the date column to datetime
df['Date'] = pd.to_datetime(df['Date'])
```

```python
type(df['Date'][0])
```

Ok date column has been updated to the correct format.

####  Checking for null values

```python
df.isnull().sum()
```

Ok we have no null values.

```python
# setting Date column to index
df.set_index('Date', inplace=True)
```

```python
df.head(2)
```

#### Visualisations

## Netflix High,Low,Open,Closed,Adj   -   From 2002 - 2021

```python
volume = df['Volume']
df.drop('Volume',1,inplace=True)
```

```python
# Plotting Data
plt.figure(figsize=(20,10))
sns.lineplot(data=df.loc['2002-05-23':'2021-09-14'], linewidth=1)
plt.title('Netflix Stocks All Time',size=28,fontweight="bold")
plt.show()
```

### Netflix High,Low,Open,Closed,Adj   -   Over 2021

```python
# plotting Data
plt.figure(figsize=(20,10))
sns.lineplot(data=df.loc['2021-01':], linewidth=1)
plt.title('Netflix Stocks - 2021',size=28,fontweight="bold")
plt.show()
```

### Netflix High,Low,Open,Closed,Adj   -  (2016 - 2021)

```python
# plotting Data
plt.figure(figsize=(20,10))
sns.lineplot(data=df.loc['2016-01':])
plt.title('Netflix Stocks (2016 - 2021)',size=20,fontweight="bold")
plt.show()
```

## Netflix Volume Stock

####  Netflix Volume Stock from 2002 to 2021

```python
# Plotting Data
plt.figure(figsize=(20,10))
sns.lineplot(data=volume.loc['2002-05-23':], linewidth=1)
plt.title('Netflix Volume All Time',size=28,fontweight="bold")
plt.show()
```

#### Volume Stocks over first year

```python
# Plotting Data
plt.figure(figsize=(20,10))
sns.lineplot(data=volume.loc['2002-05-23':'2003-05-23'], linewidth=1)
plt.title('Netflix volume stocks first year',size=28,fontweight="bold")
plt.show()
```

#### Volume Stocks 2021

```python
# Plotting Data
plt.figure(figsize=(20,10))
sns.lineplot(data=volume.loc['2021-01':], linewidth=1)
plt.title('Netflix volume stocks 2021',size=28,fontweight="bold")
plt.show()
```
